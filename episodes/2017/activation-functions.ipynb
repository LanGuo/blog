```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
xmin = -4
xmax = 4
x = np.arange(start=xmin, stop=xmax, step=.1)
```

```python
#Tanh
y = list(map(lambda x_i: np.tanh(x_i), x))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth=4)
plt.plot([xmin, xmax], [0, 0], linestyle='--', color='#888888')
plt.title('tanh', fontsize=22)
plt.show()
```

```python
#Step
def step(x):
    if x < 0:
        return 0
    if x >=0:
        return 1
```

```python
y = list(map(step, x))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth=4)
plt.plot([xmin, xmax], [0, 0], linestyle='--', color='#888888')
plt.title('step', fontsize=22)
plt.show()
```

```python
y = list(map(lambda x_i: math.pow(x_i,2), x))
```

```python
#Sigmoid
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
y = list(map(sigmoid,x))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth=4)
plt.plot([xmin, xmax], [0, 0], linestyle='--', color='#888888')
plt.title('sigmoid', fontsize=22)
plt.show()
```

```python
#Softmax
def softmax(x):
    """Compute the softmax of vector x."""
    exps = np.exp(x)
    return exps / np.sum(exps)
```

```python
y = lambda x: np.log(1+np.exp(x))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth=4)
plt.plot([xmin, xmax], [0, 0], linestyle='--', color='#888888')
plt.title('softmax', fontsize=22)
plt.show()
```

```python
def line_graph(x, y, x_title, y_title):
    plt.plot(x, y)
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.show()
```

```python
#ReLU
def relu(x):
    if x<0:
        return 0
    if x>=0:
        return x
```

```python
y = list(map(relu,x))
```

```python
plt.figure(figsize=(15,5))
plt.plot(x,y, linewidth=4)
plt.plot([xmin, xmax], [0, 0], linestyle='--', color='#888888')
plt.title('relu', fontsize=22)
plt.show()
```
